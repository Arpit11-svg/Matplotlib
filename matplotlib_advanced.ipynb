```python
df=pd.read_csv('/content/batter.csv')
df.head()
```

```python
# Task: select randomly 25 batsman and draw a scatter, between avg and strike_rate and give name to each dot
plt.figure(figsize=(15,4))
sample_df=df.sample(25)
plt.ylabel("Strike Rate")
plt.xlabel("Average")

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'],c=sample_df['runs']) #s=> size, variate size according to runs
for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
# Horizontal and vertical line accordinng to some condition
plt.figure(figsize=(18,10))
plt.ylabel("Strike Rate")
plt.xlabel("Average")
plt.axhline(130,color='red') # draw horizontal line whose strike_rate is greater than 130
plt.axvline(30,color='green') # draw vertical line whose avg is greater than 30

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs']) #s=> size, variate size according to runs

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
# Surface plot
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
xx.shape #gives 100*100 2D array
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(13,7))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
ax.set_title('Surface plot of z= x^2 + y^2')
```

```python
z=np.sin(xx)+np.sin(yy)
z.shape
```

```python
fig=plt.figure(figsize=(13,5))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.cos(xx)+np.log(yy)
fig=plt.figure(figsize=(13,5))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
# contour plotting => 3D plot into 2D plot

x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2

fig=plt.figure()
ax=plt.subplot()

p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig=plt.figure()
ax=plt.subplot()

p=ax.contourf(xx,yy,z,cmap='viridis') # contourf -> it fills the color into plot
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.sin(yy)
fig=plt.figure()
ax=plt.subplot()

p=ax.contourf(xx,yy,z,cmap='viridis') # contourf -> it fills the color into plot
fig.colorbar(p)
```

```python
# Heat map
delivery=pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
# TASK: plot heat map of that, at each ball how many sixes were played in IPL
delivery['ballnumber'].unique()
```

```python
# It have 10 values(balls) but we need only 6
temp_df=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6]) ) & (delivery['batsman_run']==6)]
```

```python
# now draw the pivot table
temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
grid=temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
plt.figure(figsize=(13,6))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21))) #delivery['overs'].unique() is replaced with list(range(1,21)) so that overs comes from [1,20]
plt.xticks(list(range(0,6)),[1,2,3,4,5,6])
plt.colorbar()
```

```python
# from above graph we can analize that most sixes played at 20th over's 3,4,6th ball.
```
