```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
batsman=pd.read_csv('/content/sharma-kohli.csv')
batsman.head()
```

```python
# 2D-line plot
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='solid',linewidth=3,marker='o',markersize=10)
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=2,marker='D',markersize=5)
plt.title("Rohit Sharma vs Virat Kohli")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='solid',linewidth=3,marker='o',markersize=10,label='RG Sharma')
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth=2,marker='D',markersize=5,label='V Kohli')
plt.title("Rohit Sharma vs Virat Kohli")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend('best') #by default location is best
plt.grid()
plt.show()
```

```python
df=pd.read_csv('/content/batsman_season_record.csv')
```

```python
df.head()
```

```python
plt.bar(df['batsman'],df['2015'],width = 0.6,color='green')
plt.show()
```

```python
# multivarient graph in single bar graph
plt.bar(np.arange(df.shape[0])-0.3,df['2015'],width=0.3,color='green',label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.3,color='blue',label='2016')
plt.bar(np.arange(df.shape[0])+0.3,df['2017'],width=0.3,color='red',label='2017')
plt.ylabel('Runs')
plt.legend()
```

```python
plt.bar(np.arange(df.shape[0])-0.3,df['2015'],width=0.3,color='green',label='2015')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.3,color='blue',label='2016')
plt.bar(np.arange(df.shape[0])+0.3,df['2017'],width=0.3,color='red',label='2017')
plt.ylabel('Runs')
plt.xlabel('Players')
plt.xticks(np.arange(df.shape[0]),df['batsman']) # 0,1,2.. replace with batsman list
plt.legend()
plt.show()
```

```python
# Problem in bar graph
colors=['red red red red red','blue blue blue blue','pink pink pink pink pink','black black black','yellow yellow yellow','green green green green green']
children=[20,39,33,19,22,44]
plt.bar(colors,children)
```

```python
# problem is of overlapping, of long names=> solution M-1: use horizontal bar / M-2: use of plt.xticks()
```

```python
colors=['red red red red red','blue blue blue blue','pink pink pink pink pink','black black black','yellow yellow yellow','green green green green green']
children=[20,39,33,19,22,44]
plt.bar(colors,children)
plt.xticks(rotation=90)
plt.show()
```

```python
# stacked bar chart
plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=df['2015']+df['2016'],label='2017')
plt.legend()
plt.show()
```
